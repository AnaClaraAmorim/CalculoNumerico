# **Transformação da base Decimal para base Binária:**
  Divide o número pela base (2) e pega os restos, até que o quociente seja igual a 0
  image.png


```python
n = 29

k = 0
denominador = 2
quociente = n // denominador
resto = n%denominador
d = []
d.append(resto)

while (quociente !=0):
  k = k+1
  n = quociente
  quociente = n // denominador
  resto = n%denominador
  d.append(resto)


print(d[::-1])
```
