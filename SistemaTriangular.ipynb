# Triangular Superior
Resolve sistema do tipo Ax = b,
em que a matriz A é tringular superior
  

Formula da resolução:

```python
def TriangularSuperior(A,b):
  n = len(b)
  x = [0]*n #cria vetor de 0 com n elementos
  x[n-1] = b[n-1]/A[n-1][n-1]

  for i in list (range(n-1,0,-1)):
    s = 0
    for j in list (range(i+1,n+1,1)):
      s = s + A[i-1][j-1]*x[j-1]

    x[i-1] = (b[i-1]-s)/(A[i-1][i-1])
  return x

#Exemplo
A = [[3,2,4],[0,(1/3),(2/3)],[0,0,1]]
b = [1,(5/3),0]
x = TriangularSuperior(A,b)
print(x)
```

# Triangular Inferior
Resolve sistema do tipo Ax = b,
em que a matriz A é tringular inferior

```python
def TriangularInferior(A,b):
  n = len(b)
  x = [0]*n #cria vetor de 0 com n elementos
  x[0] = b[0]/A[0][0]

  for i in list (range(1,n,1)):
    s = 0
    for j in list(range(1,i+1,1)):
      s = s + A[i][j-1]*x[j-1]
      
    
    x[i] = (b[i] - s)/A[i][i]
  return x

#Exemplo
A = [[1,0,0],[2,3,0],[4,5,6]]
b = [7,8,9]
x = TriangularInferior(A,b)
print(x)
```
